```python
import numpy as np

def step_function(x):
    if x>=0:
        return 1
    else:
        return 0

print("Setup complete.")
```

<b><h2>Part - 1 (OR Gate)</b>

```python
# Input
X_or=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

#Target
y_or=np.array([0,1,1,1])
```

<b><h2>Step 2 : Initialize Variables</b>

```python
weights=np.array([0,0])
bias=0
lr=0.5

print(f"Initial Weights: {weights}")
print(f"Initial Bias: {bias}")
```

<b><h2>Step 3 : Manual Calculation</b>

```python
input_sample=X_or[1]
target=y_or[1]

#1.Weighted Sum(Dot Product)
weighted_sum=np.dot(input_sample,weights)+bias
print(f"Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias}")
print(f"Sum: {weighted_sum}")

#2.Activation Function
prediction=step_function(weighted_sum)
print(f"Prediction: {prediction}")

#3.Error
error=target-prediction
print(f"Error: {error}")
```

```python
epochs=5

print("========Training========")
for epoch in range(epochs):
    print(f"\n Epoch: {epoch+1}")

    for i in range(4):
        x=X_or[i]
        y=y_or[i]

        # Predict
        weighted_sum=np.dot(x,weights)+bias
        prediction=step_function(weighted_sum)
        
        # Error
        error=y-prediction

        #Update Weights
        if error!=0:
            weights=weights+(lr*x*error)
            bias=bias+(lr*error*1)
            print(f"Mistake on [x]. Expected: {y}, Predicted: {prediction}")
            print(f"Updated Weights {x}. Expected: {y}, Predicted: {prediction}")
        else:
            print(f"No Mistake on [x]. Expected: {y}, Predicted: {prediction}")

print(f"\n ========Training Complete========")
print(f"\n Final Weights: {weights}")
print(f"Final Bias: {bias}")
```
