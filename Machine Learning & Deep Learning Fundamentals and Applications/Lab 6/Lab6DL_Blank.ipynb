# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data=load_iris()
```

```python
data
```

```python
type(data)
```

# Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  
import matplotlib.pyplot as plt      
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

# Divide the data into input and output

```python
x=data['data']
```

```python
y=data['target']
```

```python
y
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

# Display Decision Tree

```python
from sklearn import tree
```

```python
tree.plot_tree(model)
```

# Predict the x_test 

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# Save Model State

```python
import pickle
```

```python
with open('irismodel.pkl', 'wb') as f:
    pickle.dump(model, f)
```

# Load Model from disk

```python
with open('irismodel.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
model.predict(X_test)
```

# Import diabetes.csv dataset 

```python
df=pd.read_csv("diabetes.csv")
```

```python
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
df.isin(values=[0]).sum()
```

```python
col=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for col in col:
    df[col]=df[col].replace(0,np.nan)
```

```python
df
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
cols=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for col in cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
df
```

# Visualize Distributions

```python
import seaborn as sns
```

```python
cols=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for col in cols:
    sns.histplot(df[col],kde=True,color='orange')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title(f"Distribution of {col}")
```

```python
sns.pairplot(df,hue='Outcome')
```

# Convert Targer data into interger code

```python
df['Outcome']=df['Outcome'].map({'Non Diabetic':0,'Diabetic':1})
```

```python
df
```

# Divide the data into input and output

```python
x=df.iloc[ ::, :-1]
```

```python
y=df['Outcome']
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Create Model

```python
model=DecisionTreeClassifier()
```

# Fitting DecisionTreeClassifier   on dataset

```python
model.fit(X_train,y_train)
```

# Display Decision Tree

```python
tree.plot_tree(model)
```

# Predict the x_test 

```python
model.predict(X_test)
```

# Display Training Accuracy

```python
model.score(X_train,y_train)
```

# Display Test Accuracy

```python
model.score(X_test,y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,model.predict(X_test))
sns.heatmap(cm,annot=True,cmap='Blues')
plt.show()
```

# Save Model State

```python
import pickle
```

```python
with open('diabetesmodel.pkl', 'wb') as f:
    pickle.dump(model, f)
```

# Load Model from Disk

```python
with open('diabetesmodel.pkl', 'rb') as f:
    model = pickle.load(f)
```
